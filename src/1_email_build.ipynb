```python
from obspy import UTCDateTime
import math
import os


name = '.NAME Taro ZISHIN \n'
inst = '.INST Nanyang Technological University, Singapore \n'
mail = '.MAIL Nanyang Technological University, Singapore \n'
email = '.EMAIL cjcjyu597460474@gmail.com\n'
phone = '.PHONE +65 90893904 \n'
fax = '.FAX +65 90893904 \n' 
label = '.LABEL Request No.3 \n'
end = '.END \n'
blankline = '\n'

head = name+inst+mail+email+phone+fax+label+end+blankline



doc = open('../evst/catalog')
evinfo = doc.readlines()
doc.close()

try:
    os.makedirs('data')
    os.makedirs('mails')
except:
    pass

doc_file_list = open('data/file_list','w')

for i_info,info in enumerate(evinfo):

    tmp = info.split()
    ori_time = UTCDateTime(tmp[1])


    # time range  ori_time - ori_time + 3600 s
    st_time = ori_time
    ed_time = ori_time + 3600

    station = 'TSAR'
    network = 'PS'

    st_year = str(st_time.year); 
    st_mon = str(st_time.month);        st_mon=st_mon.zfill(2)
    st_day = str(st_time.day);          st_day=st_day.zfill(2)  
    st_hour = str(st_time.hour);        st_hour=st_hour.zfill(2)  
    st_min = str(st_time.minute);       st_min=st_min.zfill(2)  
    st_sec = str(int(st_time.second));  st_sec=st_sec.zfill(2)  
    
    ed_year = str(ed_time.year); 
    ed_mon = str(ed_time.month);        ed_mon=ed_mon.zfill(2)
    ed_day = str(ed_time.day);          ed_day=ed_day.zfill(2)  
    ed_hour = str(ed_time.hour);        ed_hour=ed_hour.zfill(2)  
    ed_min = str(ed_time.minute);       ed_min=ed_min.zfill(2)  
    ed_sec = str(int(ed_time.second));  ed_sec=ed_sec.zfill(2)  

    channel_number = '1'
    channel_name = '???'

    data=' '.join((station,network,st_year,st_mon,st_day,st_hour,st_min,st_sec,ed_year,ed_mon,ed_day,ed_hour,ed_min,ed_sec,channel_number,channel_name))

    # generate mail file
    filename = '%s%s%s%s%s%s'%(st_year,st_mon,st_day,st_hour,st_min,st_sec)
    doc_mail=open('mails/%s'%(filename),'w')
    doc_mail.write(head)
    doc_mail.write(data+'\n')
    doc_mail.close()

    doc_file_list.write('%s %s'%(''.join(('TSAR',st_year,st_mon,st_day,st_hour,st_min)),info))


doc_file_list.close()


```
